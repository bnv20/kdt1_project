```python
import pandas as pd
df = pd.read_pickle('/content/drive/MyDrive/L1/Notebooks/dataset/clustering_result(0_1).pkl')
```

```python
df.drop('CUSTNO',axis=1,inplace=True)
```

```python
df1 = df.copy()
df2 = df.copy()
df3 = df.copy()
df4 = df.copy()
df5 = df.copy()

```

```python
df1.labels[df1.labels!=0] = 1
df2.labels[df2.labels!=1] = 0
df3.labels[df3.labels!=2] = 0
df4.labels[df4.labels!=3] = 0
df5.labels[df5.labels!=4] = 0
```

```python
data = df.drop('labels',axis=1)
target=  df.labels
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target,
                                                    test_size=0.2, random_state=0)
```

```python
from sklearn import tree
from sklearn.tree import export_graphviz

from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
dt_tree = DecisionTreeClassifier(min_samples_leaf= 8, max_features=8,max_depth=10,max_leaf_nodes=10)
dt_tree.fit(X_train,y_train)
pred = dt_tree.predict(X_test)
plt.figure(figsize=(20,10))
tree.plot_tree(dt_tree, filled=True,
              feature_names=X_train.columns,
             
              rounded=True, fontsize=12)
plt.show()
plt.close()

# 1번그룹(cluster 0) : 남자인그룹
# 2번그룹(cluster 1) : 멤버십 미가입이면서 돈을 많이 소비한 그룹
# 3번그룹(cluster 2) : 멤버십 가입그룹
# 4번그룹 : 멤버십 가입, 미가입, 돈의 소비량도 중간/ 하지만 연령대가 있는 그룹
# 5번그룹 : 멤버십 미가입이면서 돈을 적게 소비한 그룹
```
